# Importing Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Loading the data

```python
df = pd.read_csv('C:/Users/PC/Desktop/Prodigy_Infotech_Data_Science_Internship/bar_chart_or_histogram/dataset/API_SP.POP.TOTL_DS2_en_csv_v2_427072.csv', skiprows=4)
df.head()
```

# List of aggregate keywords to filter out


```python
data_2023 = df[['Country Name', '2023']].dropna()
```

```python
aggregate_keywords = [
    'World', 'Region', 'Income', 'Aggregates', 'Countries', 'Union', 'Area', 'members', 'dividend', 'HIPC', 
    'situations', 'UN classification', 'North America', 'Euro', 'Africa', 'Asia', 'OECD'
]
```

# Function to check if a country name contains any of the aggregate keywords


```python
def is_aggregate(name):
    return any(keyword in name for keyword in aggregate_keywords)
```

# Filter out rows that contain aggregate keywords in the country name


```python
data_2023_countries_only = data_2023[~data_2023['Country Name'].apply(is_aggregate)]
```

# Sort the filtered data by population in 2023 in descending order and select the top 20 countries for better visibility


```python
data_2023_countries_only_sorted = data_2023_countries_only.sort_values(by='2023', ascending=False).head(20)
```

# Create a bar chart for the filtered data

```python
plt.figure(figsize=(12, 8))
plt.barh(data_2023_countries_only_sorted['Country Name'], data_2023_countries_only_sorted['2023'], color='skyblue')
plt.xlabel('Population in 2023')
plt.ylabel('Country')
plt.title('Top 20 Countries by Population in 2023 (Excluding Aggregates)')
plt.gca().invert_yaxis()
plt.show()
```
